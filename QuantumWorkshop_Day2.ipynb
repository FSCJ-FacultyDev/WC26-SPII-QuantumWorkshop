## More on Single-Qubit States and Gates

A quantum state can be represented in code as a state vector consisting of a pair of amplitudes which describe the contribution of each basis state.
- As we saw in the previous workshop session, quantum states are often written as kets using Dirac notation:

  |ψ⟩ = α|0⟩ + β|1)

- e.g.,

  ∣ψ⟩=(-0.34+0.43i)|0⟩+(0.68+0.49i)|1⟩

- The qubit state shown above is described as a superposition of |0> and |1>, where the complex amplitudes determine the measurement probabilities after computing their squared magnitudes.
- For a complex amplitude a+bi, the probability is a^2 + b^2.

  |-0.34+0.43i|^2 = (-0.34)^2 + (0.43)^2 = 0.1156+0.1849 ≈ 0.3005

  |0.68+0.49i|^2 = (0.68)^2 + (0.49)^2 = 0.4624 + 0.2401 ≈ 0.7025

- Final probabilities
  - P(0) ≈ 0.3005
  - P(1) ≈ 0.7025

  - Note that these sum to approximately 1 (because the original amplitudes were rounded to two places, otherwise they would sum to exactly 1), which is required for a valid quantum state.

A quantum state of a qubit can be expressed using trigonometric formulas by describing its amplitudes with angles.
- Instead of writing arbitrary complex numbers, the amplitudes are written using sine and cosine functions so that the probabilities automatically sum to 1.
- A common form (known as **Bloch sphere parameterization**) is

  |ψ> = cos(θ/2)|0> + e^(iφ) sin(θ/2)|1>

  or

  “psi equals cosine theta over two times ket zero, plus e to the i phi times sine of theta over two times ket one.”

  - (the constant e is used because it provides a convenient way to represent rotations and oscillations, fundamental to quantum mechanics).
  
  where the angles determine the magnitude and phase of the amplitudes.
- As we saw in the Day 1 workshop, these angles can be represented as a point on the Bloch sphere, allowing quantum states to be visualized as rotations on the sphere.
- We will mostly avoid this notation in this workshop; not because it isn’t important, but because saying things like “e to the i phi times sine of theta over two times ket one” before coffee can cause unnecessary stress.

## Day 2 Hands-On Exercise 1: A Bloch Sphere Visualization Tool

<img src="https://raw.githubusercontent.com/FSCJ-FacultyDev/WC26-SPII-QuantumWorkshop/main/images/bloch-sphere-visualizer.png"
     alt="Bloch Sphere Visualizer"
     width="25%">

The Bloch Sphere Visualizer at https://bloch.kherb.io/ is an interactive web tool for visualizing the state of a single qubit on a Bloch sphere.
- It provides an intuitive way to see how quantum states change when quantum gates or rotations are applied.
### Description
- Z axis (top and bottom): Represents the computational basis states |0⟩ and |1⟩, which determine the measurement outcomes.
- X axis (left to right across the equator): Represents equal superpositions of |0⟩ and |1⟩ with no phase difference, such as (|0⟩±|1⟩)/√2.
- Y axis (front to back across the equator): Represents equal superpositions where the amplitudes differ by an imaginary phase (±i), such as (|0⟩±i|1⟩)/√2.
### Demonstration
- Rotation Around the X Axis
  - When you rotate around the X axis, the state moves in a vertical circle that passes through the Y direction.
  - Think of it like spinning the sphere around the horizontal rod that goes left to right.
- Start State (INIT)
  - Start at |0⟩, which is the north pole.
- Apply an X-axis rotation of +90° (click +90° on the X axis)
  - The state moves from the north pole toward the front of the sphere, landing on the +Y axis.
    - Mathematically, this is approximately (|0⟩ + i|1⟩)/√2, a superposition with phase where the probabilities of measuring 0 or 1 are equal: P(0) = 0.5 and P(1) = 0.5.
    - This introduces an **i phase**, meaning the amplitude of |1⟩ is multiplied by i (i.e., √-1), creating a 90° phase difference between the amplitudes of |0⟩ and |1⟩.
    - This phase difference places the state on the +Y axis of the Bloch sphere.
    - This does not change measurement probabilities, but it affects how the state interferes with other states during computation.
- Rotate another +90° on the X axis (for a total 180°). Now the state reaches the south pole, or |1⟩
  - The 180° rotation around X flips the qubit (swaps the amplitudes of |0⟩ and |1⟩.
  - This is equivalent to the behavior of a quantum NOT gate (e.g., the **Pauli-X gate**), which flips the state of a qubit.
- Continue rotating another +90°
  - The state becomes (|0⟩ - i|1⟩)/√2, a superposition with equal probabilities but a -90° phase shift.
  - Now we are on the opposite side of the Y axis.
- Rotate a final +90° (360° total), returning to |0⟩.
- In general:
  - Rotation around the X axis mixes amplitudes and introduces phase (flips the qubit through superposition).
  - Rotation around the Y axis mixes amplitudes without introducing complex phase (moves the qubit smoothly between |0⟩ and |1⟩).
  - Rotation around the Z axis changes phase only (spins the phase around the equator while probabilities stay the same).
- Pauli Gates are basic single-qubit quantum gates that perform 180° rotations of a qubit’s state around the principal axes of the Bloch sphere.
  - Pauli-X (X): 180° rotation around the X axis (swaps |0⟩ and |1⟩, similar to a classical NOT gate)
  - Pauli-Y (Y): 180° rotation around the Y axis (swaps |0⟩ and |1⟩ while introducing phase)
  - Pauli-Z (Z): 180° rotation around the Z axis (changes the relative phase between |0⟩ and |1⟩ without changing measurement probabilities)

## Single Qubit with Example Amplitudes

- These amplitudes determine the probabilities of measurement and can be represented in algebraic form within the table.
- Complex numbers can be visualized on the complex plane, where the horizontal axis represents the real part and the vertical axis represents the imaginary part.
- On this plane, the amplitude’s magnitude corresponds to its distance from the origin, and its direction corresponds to its phase (the angle from the real axis).
- The magnitude determines measurement probability (by squaring it, as we saw above), while the phase  influences how states interfere (combine and interact) during computation.

  - => if the probability of outcome 0 is p, then the probability of outcome 1 is 1 - p.

### Polar Form
- A complex number can be represented by a magnitude r and an angle θ (which represents the direction).
- Using this information, we can write the number z = x + iy as
  - x = r cosθ
  - y = r sinθ
- which gives us
  - z = r(cos θ + i * sin θ)
- This is useful in quantum computing because quantum gates often rotate amplitudes by changing their phase θ while keeping their magnitude the same.
  - Many quantum operations are rotations, and rotations are much easier to describe with magnitude + angle than with real/imaginary parts.



### Encoding Single-Qubit States with Python
- A single-qubit quantum state can be expressed as a Python list of two amplitudes, where the index of an amplitude is the corresponding outcome.
- Python represents the imaginary unit, i, in a complex number with the symbol j.
  - j is used instead of i because in engineering i usually represents electric current; Python's creator (Guido van Rossum) used this convention as a design choice.
- To encode the state described above:
```
  state = [0.2958+0.51235j, -0.40311+0.69821j]
```

- The index of each amplitude is the corresponding outcome.

## The Hadamard Gate
Recall the Hadamard gate, which places a qubit into an equal superposition of the states |0> and |1>.
- Applied to |0> the Hadamard produces
  - H|0> = (|0> + |1>) / √2
- and applied to |1> it produces
  - H|1> = (|0> - |1>) √2
### Why √2?
Because probability=∣amplitude∣^2
- The sum of the squares of the amplitudes must equal 1, since those squares represent measurement probabilities.

Applying the Hadamard gate to each qubit in a register creates a superposition of all possible basis states.
  - For example, applying Hadamard gates to three qubits creates a uniform superposition over the eight states 000 through 111, each initially having equal amplitude.
- This step is commonly used at the beginning of many quantum algorithms to represent all possible candidate solutions simultaneously.

## Multi-Qubit Systems and Entanglement

When we move from a single qubit to multi-qubit systems, the state space grows exponentially: two qubits can represent four basis states, three qubits represent eight, and so on.
- Unlike classical bits, qubits can become entangled, meaning their states cannot be described independently of one another.
  - this is where quantum computing stops being “probability with fancy math” and becomes fundamentally different from classical systems.

### Separable Qubits

- In classical systems, if you have two bits:

    bit A = 0

    bit B = 1

  you can fully describe the system by describing each bit separately.
- Even if they are correlated (for example, always opposite), you can still describe them independently:

    “B is the opposite of A.”

- Two qubits are **separable** when each one has its own individual state, and the overall system is just those two states considered together.

  Qubit A has its own condition.

  Qubit B has its own condition.

- Nothing special ties them together beyond being part of the same system.
- For example, if one qubit is in a superposition and the other is just in state 0, there is no deep connection between them — knowing everything about one tells you nothing new about the other.

### Entangled State
- In an entangled system, two qubits become linked so their measurement results are correlated.
  - For example, a pair might always produce matching results — both 0 or both 1 — even though neither qubit has a definite value before measurement.
- Entangled qubits cannot be described independently; only the combined system fully describes the state.
  - This allows quantum algorithms to coordinate multiple qubits and manipulate information stored in the relationships between them.
 - Quantum algorithms using entanglement can manipulate many possible states together and solve certain problems more efficiently than classical methods.



## Circuit-Based Quantum Computing Model
This model describes quantum computation as a sequence of quantum gates applied to qubits, similar to how classical circuits apply logic gates to bits.
- Qubits are initialized in a known state, transformed step-by-step using unitary operations (quantum gates), and finally measured to produce classical results.
- Each gate acts on one or more qubits and changes the overall quantum state according to linear algebra.
- The circuit-based model provides a structured and visual way to design algorithms, making it the most widely used framework for building and understanding quantum programs today.
