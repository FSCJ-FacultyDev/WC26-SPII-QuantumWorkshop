## Quantum Oracles

- In quantum computation, an oracle is used to recognize specific outcomes and mark them during a computation.
  - We call the outcomes we want to identify “good” outcomes, while all others are “bad” outcomes.
- Oracles typically mark good outcomes in one of two ways:
  - Phase oracle: Rotates the amplitudes of the good outcomes by 180°, effectively flipping their phase.
  - Bit oracle: Entangles outcomes with an additional qubit, so good outcomes correspond to measuring 1 and bad outcomes correspond to measuring 0.
- These actions mark the states we care about without directly measuring them (because measurement would collapse the quantum state and destroy the superposition).
- The internal implementation of an oracle is independent of how it is later used, which is why it is called an “oracle.”
- A phase oracle can be easily constructed from a bit oracle, since the bit oracle already marks the good outcomes.

## Shot-Based Execution and Statistics
Shot-based execution refers to running the same quantum circuit multiple times, where each individual run (a “shot”) produces a single measurement outcome.
- Because quantum measurement is probabilistic, one execution does not reveal the full state of the system.
- Instead, meaningful results emerge by collecting statistics across many shots and analyzing the distribution of outcomes.
- This statistical approach is fundamental to interpreting quantum computations and connecting amplitudes to observable probabilities.

## Sources of Noise in Quantum Systems
Noise in quantum systems refers to environmental and technical disturbances that disrupt fragile quantum states.
- Because qubits must remain isolated to preserve superposition and entanglement, interactions with heat, electromagnetic radiation, material defects, or imperfect control signals can introduce errors.
- These disturbances lead to decoherence (loss of quantum state) and gate inaccuracies, reducing computational reliability.
- Understanding and mitigating noise is essential for improving quantum hardware performance and enabling scalable quantum computation.


## Error Mitigation
Error mitigation is used to reduce the impact of noise and imperfections in quantum computations without requiring full error correction.
- Because current quantum hardware is susceptible to decoherence, gate errors, and measurement inaccuracies, mitigation methods aim to improve result accuracy through strategies such as repeated sampling, calibration adjustments, and post-processing corrections.
- These approaches do not eliminate errors entirely but help extract more reliable information from noisy quantum devices, making them especially important for near-term quantum systems.
