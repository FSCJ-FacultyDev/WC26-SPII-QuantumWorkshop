# Essential Quantum Mechanics

**‚ÄúI think I can safely say that nobody understands quantum mechanics.‚Äù**  
‚Äî Richard Feynman (1965)

Quantum mechanics principles are essential for constructing quantum hardware, but they are not the primary focus when learning quantum computing concepts.
- Instead, we‚Äôll focus on simple models and algorithms that help explain how quantum information behaves, without getting into the physics.

While you don‚Äôt need deep physics to understand the basics of quantum computers, the following ideas can help explain why quantum computing behave differently from classical ones:

1. **Superposition:** a quantum system can exist in multiple states at the same time (e.g., both 0 and 1). This is what allows quantum computers to process many possibilities simultaneously.

   ‚ñ∂ [Watch: Superposition Explained](https://www.youtube.com/watch?v=IHDMJqJHCQg)

2. **Measurement:** observing a quantum system forces it to ‚Äúchoose‚Äù one outcome from many possibilities. This is why quantum results are inherently uncertain, which is why programs must be run many times.

3. **Entanglement:** Two or more qubits can become linked so that the state of one is related to the state of another. This enables coordinated multi-qubit behavior and is essential for real quantum computation.

   ‚ñ∂ [Watch: Entanglement Explained](https://www.youtube.com/watch?v=rqmIVeheTVU)

4. **Probability Distributions:** Quantum results follow probability patterns rather than producing a single fixed answer. Results are interpreted statistically, not as one ‚Äúcorrect‚Äù output.

5. **Noise:** Real quantum systems are affected by environmental interference, which introduces errors. This is why quantum results are imperfect and why error mitigation is an important area of research.

#Quantum Computing Terminology

1. **Quantum Advantage:** The performance benefit quantum computers can achieve over classical computers for certain problem types, such as simulating quantum systems or performing specific searches more efficiently.

2. **Quantum Parallelism:** The ability of a quantum computer to apply one operation to many possible values at the same time, similar to an unlimited SIMD (Single Instruction, Multiple Data) model, enabling faster computation.

![SIMD](https://raw.githubusercontent.com/FSCJ-FacultyDev/WC26-SPII-QuantumWorkshop/main/images/simd.png)

  - One operation acts on many possibilities at once, like SIMD.
  - **But** you never see the intermediate values.
  - The computation reshapes a probability distribution using interference:
    - Good answers are amplified
    - Bad answers are suppressed
  - Measurement gives you one sample, not the full state.

3. **Shot:** A single execution of a quantum circuit that produces one  outcome. Because results are probabilistic, many shots are run to estimate the full probability distribution of possible outcomes.

4. **Quantum State:** A mathematical description of a quantum system that encodes all possible outcomes of a computation, along with their probabilities and relative phase (direction) information.

5. **Measurement:** The process of observing a quantum state, which produces one probabilistic outcome from many possibilities and requires repeated runs (shots) to understand the full distribution.

6. **Amplitude:** A number that describes how strongly a quantum state contributes to the overall system, and its square determines the probability of observing that state when measured.

7. **Butterfly Computational Pattern:** A structure of quantum operations where pairs of values are repeatedly recombined in parallel to shape the final probability distribution before measurement.

8. **Qubit (Quantum Bit):** The basic unit of information in a quantum computer. Unlike a classical bit that is either 0 or 1, a qubit can exist in a superposition of both states at the same time, and can also become entangled with other qubits to enable powerful quantum computations.

9. **Quantum Superposition:** The ability of a quantum system to exist in multiple states at the same time, leading to uncertainty in measurement outcomes.

10. **Quantum Entanglement:** A property where qubits become linked so that their states cannot be described independently, making their measurement results interdependent.

11. **Measurement Uncertainty**: The inherent unpredictability of which outcome will be observed when measuring a quantum system in superposition.

##Why Quantum?

Quantum computing offers speedups beyond classical scaling for select problems (e.g., Fourier transforms, search, simulation). This is known as the **quantum advantage**.

![QuantumVsClassical](https://raw.githubusercontent.com/FSCJ-FacultyDev/WC26-SPII-QuantumWorkshop/main/images/quantum-vs-classical.png)

Computations that are impractical or impossible classically (quantum physics, chemistry, true randomness) are significantly faster
- e.g., in chemistry, predicting a molecule‚Äôs energy levels requires modeling how many interacting electrons behave under quantum mechanics.
  - Each additional electron doubles the size of the system‚Äôs state space, causing classical simulations to scale exponentially in time and memory.
- Even powerful classical supercomputers struggle with accurate simulations for moderately sized molecules (beyond a few dozen interacting electrons).
  - A quantum computer represents electron states directly using qubits, allowing it to naturally model superposition and entanglement.
- Quantum algorithms can estimate molecular energy levels with far fewer resources, enabling applications such as drug discovery, catalyst design, and materials science.

![MoleculeCost](https://raw.githubusercontent.com/FSCJ-FacultyDev/WC26-SPII-QuantumWorkshop/main/images/molecule-cost.png)




In addition to speedups and reduction of resources, quantum computers can perform some computations that classical computers cannot, such as
- **quantum modeling:** They can simulate molecular behavior at the quantum level, which is useful for applications such as drug discovery and energy technologies.
- **true randomness:** Quantum systems produce inherently random outcomes, unlike classical systems that rely on deterministic algorithms or pseudo-random number generators.
  - This is especially valuable for cryptography and security applications

## Key Characteristics

- Quantum parallelism: single operations act on many value pairs simultaneously (SIMD-like)
- Parallel updates can cause unwanted side effects, requiring careful algorithm design
- Measurement is probabilistic: one run gives one outcome from a distribution
- Repeated runs (‚Äúshots‚Äù) reveal probability distributions, not deterministic answers
- Results are read as binary strings, enabling digital or probabilistic interpretations

# Demonstrating Quantum Parallelism

We can demonstrate quantum parallelism by swapping odd and even components of an 8-element array.

<img src="https://raw.githubusercontent.com/FSCJ-FacultyDev/WC26-SPII-QuantumWorkshop/main/images/table1-1-probability-pairs.png"
     alt="Probability Pairs"
     width="50%">


For a classical implementation, we can write out all the operations or use a for loop to execute sequentially (Fig 1.2)
- the number of operations necessary will grow linearly with
the number of swaps we want to perform.

<img src="https://raw.githubusercontent.com/FSCJ-FacultyDev/WC26-SPII-QuantumWorkshop/main/images/sequential-swap.png"
     alt="Sequential Swap"
     width="75%">

```python
# explicit loop-based swap

values = [0.05, 0.11, 0.13, 0.02, 0.34, 0.17, 0.06, 0.12]

print('before swap:', values)
# swap adjacent elements (0‚Üî1, 2‚Üî3, 4‚Üî5, 6‚Üî7)
for i in range(0, len(values), 2):
    values[i], values[i+1] = values[i+1], values[i]

print('after swap:',values)
```

The code shown above is an example of an algorithm which uses explicit, conventional iteration (i.e., a loop).

We can use Python slicing to demonstrate a more transparent version, which  removes the visible loop.



```python
# slice-based swap

values = [0.05, 0.11, 0.13, 0.02, 0.34, 0.17, 0.06, 0.12]
print('before swap:', values)
values[::2], values[1::2] = values[1::2], values[::2]
print('after swap:',values)
```

This version appears to swap all adjacent elements at once, but the underlying implementation still performs conventional iteration
- It is simply abstracted away in optimized C code.

- The loop has not disappeared; it is just hidden.

Quantum state evolution is fundamentally different.

We treat the state as one mathematical object that gets transformed in a single step.

<img src="https://raw.githubusercontent.com/FSCJ-FacultyDev/WC26-SPII-QuantumWorkshop/main/images/quantum-swap.png"
     alt="Quantum Swap"
     width="50%">

Instead of updating individual entries one-by-one, the entire state vector is multiplied by a transformation matrix.
- The operation applies to all possibilities simultaneously because they are part of the same mathematical object.
- In classical computing, even when iteration is hidden, individual elements are still processed independently.
- In quantum computing, there are no independent values being iterated over; there is a single state evolving as a whole.

## Quantum Algorithms

- Programs evolve a quantum state across exponentially many outcomes in parallel (**quantum parallelism**)
  - Instead of checking each possibility one by one, quantum algorithms use simple repeated steps that mix the possibilities together.
- They use linear transformations that recombine values in repeating ‚Äúbutterfly‚Äù patterns

  <img src="https://raw.githubusercontent.com/FSCJ-FacultyDev/WC26-SPII-QuantumWorkshop/main/images/butterfly-pattern.png"
     alt="Butterfly Pattern"
     width="50%">

- Think of this as stirring ingredients in a bowl; each simple step mixes everything together a little more, gradually increasing the amount of the ‚Äúright‚Äù answer and reducing the rest.
  - After this mixing process shapes the probabilities, measurement produces one outcome.
  - The solution is not in a single run, but in the statistical pattern that appears across many runs.

- Note that in the stirring analogy above, we do not know the right answer ahead of time; the algorithm is designed so that the structure of the problem causes the correct answers to amplify automatically.
  - A quantum algorithm doesn‚Äôt ‚Äúguess and boost.‚Äù Instead:
  - The problem is encoded into a function (often called an **oracle**).
  - That function marks correct solutions in a mathematically detectable way.
  - The quantum circuit is built so that interference increases the amplitudes of those marked states.
  - We are not manually choosing which answer to amplify; the problem structure determines it.

## Interference does the work

Constructive interference increases amplitudes of states that satisfy the condition, while destructive interference suppresses the others through repeated linear transformations (the butterfly recombinations).

## Example: Grover‚Äôs Algorithm
- Suppose one item in a database satisfies a condition, but we do not know which one.
- The algorithm begins in an equal superposition, meaning every possible input starts with the same amplitude (and therefore the same probability).
- A quantum circuit (an oracle) is designed to evaluate the condition f(x).
- When applied to the superposition, it flips the phase of any state that satisfies the condition.
- The function is applied to all possible inputs at once, without checking them individually.
- Repeated mixing steps then use interference to increase the amplitude of the state that satisfies the condition.
- When the system is measured, the correct item appears with high probability.

- Common quantum algorithm patterns:
  - Sampling from hard-to-generate distributions
  - Searching for digital answers (e.g., Grover, Shor)
  - Estimating probabilities via many shots (analog-style results)
  - Emphasis on hardware- and library-agnostic thinking as the ecosystem evolves

### 2. Visualizing Superposition
- The following code creates a simple plot of various qubit amplitudes.
  - Each bar represents the amplitude associated with the basis states ‚à£0‚ü© and ‚à£1‚ü© (the basis states are the basic reference states used to describe any qubit).
  - By visualizing these amplitudes, we can see how different quantum states are represented mathematically and how their amplitudes relate to measurement probabilities.

- |0‚ü© (Ground State)
  - The bar for |0‚ü© is 1.
  - The bar for |1‚ü© is 0.
  - This means the qubit is definitely in state 0.
  - If we measure it, we will always get 0.
  - This behaves exactly like a classical bit set to 0.
- |1‚ü© (Excited State)
  - The bar for |1‚ü© is 1.
  - The bar for |0‚ü© is 0.
  - The qubit is definitely in state 1.
  - If we measure it, we will always get 1.
  - This behaves like a classical bit set to 1.
- |+‚ü© = (|0‚ü© + |1‚ü©)/‚àö2 (Equal Superposition)
  - Both bars are the same height.
  - The qubit is in a balanced combination of 0 and 1.
  - If we measure it many times, we will get 0 about 50% of the time and 1 about 50% of the time.

**NOTE:**
The qubit is not secretly 0 or 1 ‚Äî it genuinely has both possibilities until measurement.

- |‚àí‚ü© = (|0‚ü© ‚àí |1‚ü©)/‚àö2 (Equal Superposition with Phase)
  - The bars have the same magnitude, but one is negative.
  - The probabilities are still 50% for 0 and 50% for 1.
  - The only difference is the sign (phase) of one amplitude.
  - Even though measurement probabilities are the same as |+‚ü©, this state behaves differently when we apply additional quantum operations because the negative sign affects interference.

- The height of each bar represents amplitude.
- The square of the height determines probability.
- The sign (positive or negative) affects interference, even if it does not change the immediate measurement probability.

## Why Do Quantum Computers Look So Weird?
(Post-Quantum, n.d.)

- Quantum computers resemble copper chandeliers because their design supports the extreme cooling and shielding required for superconducting qubits.
- Cryogenic cooling to millikelvin temperatures (20‚Äì100 mK) is essential for superconducting qubits; even tiny amounts of thermal energy cause decoherence (the loss quantum state from environmental disturbance) and computational errors.
- At near-absolute-zero temperatures, materials become superconducting with zero electrical resistance, which enables key quantum computing techniques and preserves coherence.
- Suspending quantum computers from the ceiling reduces ground-borne mechanical vibrations that could disturb fragile qubit states, creating a more stable operating environment.
- A "chandelier-style" vertical design improves thermal isolation, maintains ultra-low temperatures, and allows efficient organization of cryogenic stages and wiring for easier maintenance.
